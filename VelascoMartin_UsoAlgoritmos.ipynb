```python
#TESTEABLE
class SeleccionFutbol:
    

    def mochilaParaJugadores(jugadores):
        maxP, maxD, maxM, maxDel = 3, 5, 10, 5
        n = len(jugadores)

        tabla = [[[[[0 for _ in range(maxDel + 1)] for _ in range(maxM + 1)]
                   for _ in range(maxD + 1)] for _ in range(maxP + 1)]
                   for _ in range(n + 1)]

        nombreAposicion = {
            "Portero": (1, 0, 0, 0),
            "Defensa": (0, 1, 0, 0),
            "Mediocampista": (0, 0, 1, 0),
            "Delantero": (0, 0, 0, 1)
        }

        for i in range(1, n + 1):
            nombre, posicion, valor = jugadores[i - 1]
            p_i, d_i, m_i, f_i = nombreAposicion[posicion]

            for p in range(maxP + 1):
                for d in range(maxD + 1):
                    for m in range(maxM + 1):
                        for f in range(maxDel + 1):
                            tabla[i][p][d][m][f] = tabla[i - 1][p][d][m][f]
                            if p >= p_i and d >= d_i and m >= m_i and f >= f_i:
                                valor_con = tabla[i - 1][p - p_i][d - d_i][m - m_i][f - f_i] + valor
                                tabla[i][p][d][m][f] = max(tabla[i][p][d][m][f], valor_con)

        i, p, d, m, f = n, maxP, maxD, maxM, maxDel
        elegidos = []

        while i > 0:
            if tabla[i][p][d][m][f] != tabla[i - 1][p][d][m][f]:
                nombre, posicion, valor = jugadores[i - 1]
                elegidos.append((nombre, valor, posicion))
                p_i, d_i, m_i, f_i = nombreAposicion[posicion]
                p -= p_i
                d -= d_i
                m -= m_i
                f -= f_i
            i -= 1

        return elegidos[::-1]

    
    def quicksort(secuencia, key_function, reverse=False):
        def funcion(x):
            return -key_function(x) if reverse else key_function(x)
        return SeleccionFutbol._quicksort(secuencia, funcion)

   
    def _quicksort(secuencia, key_function):
        if len(secuencia) <= 1:
            return secuencia
        pivote = secuencia[0]
        izquierda, derecha = SeleccionFutbol.particion(secuencia[1:], pivote, key_function)
        return SeleccionFutbol._quicksort(izquierda, key_function) + [pivote] + SeleccionFutbol._quicksort(derecha, key_function)

   
    def particion(secuencia, pivote, key_function):
        izquierda, derecha = [], []
        for elem in secuencia:
            if key_function(elem) <= key_function(pivote):
                izquierda.append(elem)
            else:
                derecha.append(elem)
        return izquierda, derecha

    def organizarScouting(grafo,ciudadInicial=None):
        
        if ciudadInicial is None:
            ciudadInicial = list(grafo.keys())[0]

        arbol = {x: dict() for x in grafo.keys()}
        vistos = set()
        candidatos = {x: (None, float("inf")) for x in grafo.keys()}
        vistos.add(ciudadInicial)

        while len(vistos) < len(grafo):
            for adyacente, peso in grafo[ciudadInicial].items():
                if adyacente not in vistos and peso < candidatos[adyacente][1]:
                    candidatos[adyacente] = (ciudadInicial, peso)

            mejor = min(candidatos, key=lambda x: candidatos[x][1])
            origen = candidatos[mejor][0]
            arbol[mejor][origen] = candidatos[mejor][1]
            arbol[origen][mejor] = candidatos[mejor][1]
            ciudadInicial = mejor
            vistos.add(mejor)
            candidatos.pop(mejor)

        return arbol

    def rutasParaScouter(arbol,ciudadScouter):

        if ciudadScouter not in arbol:
            print(f" La ciudad '{ciudadScouter}' No esta disponible en nuestra red de ojeadores.")
            return
    
        destinos = list(arbol[ciudadScouter].keys())
        print(f" El scouter en {ciudadScouter}  tiene que ir a: {destinos}")
    


```

## 1. Complejidad temporal de mis implementaciones

### 1.1. Función `mochilaParaJugadores`

Para este método he implementado una variante del problema de la **mochila 0/1**, pero añadiendo restricciones por tipo de jugador: porteros, defensas, mediocampistas y delanteros. En lugar de usar una mochila con solo peso y valor, he usado una tabla de cinco dimensiones donde las últimas cuatro dimensiones representan las cantidades maximas permitidas de cada tipo de jugador (por ejemplo, maxP = 3 porteros, maxD = 5 defensas, etc.).

Esto hace que, como me explicaste en tutoria, aunque hay 4 bulces for anidados, como estos se van recorrer siempre el mismo numero de veces la complejidad temporal de mi algoritmo no es de orden O(n^5) sino que es del orden:

`O(n x maxP x maxD x maxM x maxDel)` mas concretamente = `O(n x 3 x 5 x 10 x 5)` es decir `O (n x 750)`

donde n es el número total de jugadores.

### 1.2. Función `quicksort`

En mi código he implementado una versión personalizada del **algoritmo Quicksort**, donde separo la lógica en dos funciones: una principal (quicksort) que se encarga de aplicar la función clave y del orden ascendente o descendente, y otra auxiliar (_quicksort) que es la que realiza la recursión.

La complejidad  de mi implementación es `O(n log n)`, ya que, al igual que en el Quicksort clásico, selecciono un pivote (en mi caso, el primer elemento de la lista) y divido el resto de elementos en dos listas: izquierda y derecha, dependiendo de si son menores o mayores que el pivote según la función key_function. Esta división se repite recursivamente sobre cada sublista.

Lo que hace que mi implementación mantenga esa complejidad es que en cada nivel de recursión divide los datos y hace comparaciones lineales `O(n)`, y como esto ocurre log n veces, el coste total es `O(n log n)`.

### 1.3. Función `organizarScouting`

Para organizar el scouting entre ciudades, he representado las conexiones como un grafo con pesos y he utilizado el **algoritmo de Prim** para construir un árbol de expansión mínima. Lo que hago es partir desde una ciudad e ir conectando las ciudades restantes eligiendo siempre la opción más barata en cuanto a peso (distancia o coste).

Mi implementación tiene una complejidad de `O(V^2)`, ya que en cada iteración recorro los vecinos de la ciudad actual y busco el mejor candidato con min(), lo cual me resulta más intuitivo aunque no sea lo más eficiente.

## 2. Por qué he elegido estos algoritmos

 ### 2.1. Elección de programacion dinamica para la mochila

He elegido resolver el problema de selección de jugadores usando **programacion dinamica** porque me aseguraba encontrar la solucion óptima cumpliendo todas las restricciones que yo mismo había definido. Es decir , no podía pasarme de 3 porteros, 5 defensas, 10 mediocampistas ni 5 delanteros. Esas condiciones complicaban mucho el problema, y con programación dinámica puedo construir una tabla que tenga en cuenta todas esas condiciones a la vez.

Además, quería que la solución fuera exacta, no solo una aproximación o una buena suposición. Usar una matriz de cinco dimensiones (tabla[i][p][d][m][f]) me ha permitido controlar perfectamente cada combinación posible de jugadores seleccionados hasta ese punto. Aunque es costoso en espacio, he considerado que el problema era manejable porque el número de jugadores no va a ser enorme en mi contexto.

## 3. Justificación de las estructuras de datos

### 3.1. Tabla multidimensional en la mochila

Para la función de seleccion de jugadores, he implementado una **tabla de cinco dimensiones** que representa el estado del problema en cada paso de la programación dinámica. Las dimensiones son:  

- el número de jugadores procesados hasta el momento,  
- y la cantidad de jugadores seleccionados en cada una de las posiciones: porteros, defensas, mediocampistas y delanteros.

Aunque pueda parecer una estructura excesivamente grande, en este caso no supone un problema porque los valores máximos para cada posición son bastante reducidos (por ejemplo, hasta 3 porteros, 5 defensas, etc.). Gracias a eso, el uso de memoria es perfectamente asumible.


### 3.2. Listas para Quicksort

Para ordenar las listas de jugadores según un criterios, he implementado una versión propia del algoritmo Quicksort, usando listas estándar de Python.  

Divido la secuencia original en sublistas izquierda y derecha en cada paso recursivo, según una función de clave (key_function) que puedo personalizar.

Escogí **listas** normales porque es lo más natural y directo en Python: son estructuras dinámicas y adecuadas para manipular secuencias. Además, el código queda más limpio y legible, sin necesidad de estructuras adicionales como pilas o colas.

### 3.3. Diccionarios para el grafo

Para representar el grafo que utilizo en la parte de scouting (es decir, las conexiones entre ciudades), he optado por utilizar **diccionarios de diccionarios**. Cada ciudad es una clave, y su valor es otro diccionario que contiene las ciudades vecinas y el peso (o coste) del camino hacia ellas.

Este tipo de estructura me resulta muy util porque:

- es muy fácil de modificar y mantener,
- y se adapta bien al algoritmo que he implementado, que es similar a Prim para construir un árbol de expansión mínima.

También he utilizado un **diccionario** llamado candidatos para guardar el mejor camino provisional hacia cada ciudad que aún no ha sido visitada. Esto me permite seleccionar de manera eficiente la siguiente ciudad en cada iteración, manteniendo el coste del grafo lo más bajo posible.
